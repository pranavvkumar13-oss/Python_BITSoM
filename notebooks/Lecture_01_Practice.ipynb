## C. Numbers & Arithmetic


```python
a = 15
b = 4

print('Addition:', a + b)
print('Subtraction:', a - b)
print('Multiplication:', a * b)
print('Division:', a / b)
print('Floor division:', a // b)
print('Modulus:', a % b)
print('Exponentiation:', 2 ** 6)
```

## G. Formatting & Padding


```python
print('7'.zfill(4))
template = 'Hello, {} {}'
print(template.format('Ada', 'Lovelace'))
print('{:>10}'.format('right'))
print('{:^10}'.format('center'))
```

```python
first, last = 'Ada', 'Lovelace'
print(f'Hello, {first} {last}')
print(f'{3.14159:.2f}')
```

## K. Mini Tasks


```python
raw_code = '  AB-123  '
second_code = 'xy 9'
third_code = 'A__B  77'
# remove spaces from both ends, convert to uppercase, and remove hyphens
print(raw_code.strip().upper().replace('-', ''))
# Remove spaces and convert to uppercase
print(second_code.strip().replace(' ', '').upper())
# replace double underscores with single underscore and double spaces with single space
print(third_code.strip().replace('__', '_').replace('  ', ' ').strip())
```
